```python
import pandas as pd

old_df = pd.read_csv('data.csv')
old_df
```

```python
new_df = pd.read_csv('new_data.csv', index_col="ID")
new_df.pop("release_date")
new_df
```

```python
years = new_df["year"].unique()
years
```

```python
decade = 20
tmp_df = pd.DataFrame()
for year in years:
    if year % 10 != 0: 
        tmp_df = tmp_df.append(new_df.loc[new_df["year"] == year])
    else:
        if decade == 0:
            tmp_df.to_csv(f"data_by_decade/data_from_0{decade}s.csv")
        else:
            tmp_df.to_csv(f"data_by_decade/data_from_{decade}s.csv")
        tmp_df = pd.DataFrame()
        decade += 10
        if decade >= 100:
            decade = 0
```
