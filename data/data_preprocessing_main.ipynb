```python
import pandas as pd
import numpy as np

original_df = pd.read_csv('data_original.csv')
important_features =  ['valence',
 'year',
 'acousticness',
 'artists',
 'danceability',
 'duration_ms',
 'energy',
 'explicit',
 'instrumentalness',
 'liveness',
 'loudness',
 'mode',
 'popularity',
 'speechiness',
 'tempo']

important_df = original_df[important_features]

important_df
```

```python
years = important_df["year"].unique()

frames_old = []
frames_new = []

for year in years:
    important_df_year = important_df.loc[important_df['year'] == year]

    print(year," avg: ",important_df_year["popularity"].mean(), " min: ",important_df_year["popularity"].max())

    old_df_with_low_popularity = important_df_year.loc[ (important_df_year["popularity"].mean() < 25)  &                                                                       (important_df_year["popularity"] < 40)  ]
    new_df_or_old_favorites  = important_df_year.loc[ (important_df_year["popularity"].mean() >= 25)  |                                                                      (important_df_year["popularity"] >= 40)  ]
    frames_old.append(old_df_with_low_popularity)
    frames_new.append(new_df_or_old_favorites)

old_dfs = pd.concat(frames_old)
new_dfs = pd.concat(frames_new)
```

```python
#Split on decades (our first choise)
decade = 20
tmp_df = pd.DataFrame()
ALL_dfs = {}
for year in years:
    if year % 10 != 0: 
        tmp_df = tmp_df.append(important_df.loc[important_df["year"] == year])
    else:
        if decade == 0:
            #tmp_df.to_csv(f"data_by_decade/data_from_0{decade}s.csv")
            ALL_dfs[decade]=tmp_df
        elif year == 2020:
            tmp_df = tmp_df.append(important_df.loc[important_df["year"] == year])
            #tmp_df.to_csv(f"data_by_decade/data_from_{decade}s.csv")
            ALL_dfs[decade]=tmp_df
            break
        else: 
            #tmp_df.to_csv(f"data_by_decade/data_from_{decade}s.csv")
            ALL_dfs[decade]=tmp_df
        tmp_df = pd.DataFrame()
        decade += 10
        if decade >= 100:
            decade = 0
            
        tmp_df = tmp_df.append(important_df.loc[important_df["year"] == year])
```

```python
#Normalize old and bad data in old_dfs

old_popularity = np.array(old_dfs["popularity"].values)
old_popularity += 3
coefficient = 100 / old_popularity.max()
old_popularity = (old_popularity * coefficient).astype(int)
old_dfs["popularity"] = old_popularity
old_dfs
```

```python
# Here normalize our data for every decade
for dec in ALL_dfs.keys():
    decade_popularity = np.array(ALL_dfs[dec]["popularity"].values)
    temp_popularity = np.array([])
    if( decade_popularity.mean() < 25 ):
        for pop in decade_popularity:
            pop +=  (decade_popularity.max() - pop) / 4
            temp_popularity = np.append(temp_popularity, pop)
    else:
        temp_popularity = decade_popularity
    coefficient = 100 / temp_popularity.max()
    decade_popularity = (temp_popularity * coefficient).astype(int)
    ALL_dfs[dec]["popularity"] = decade_popularity
    print(ALL_dfs[dec]["popularity"].mean())
```

```python
#Change artists categorical values into numerical for using it in sklearn classifiers
from sklearn import preprocessing

lab_enc = preprocessing.LabelEncoder()

old_artists = old_dfs["artists"]
encoded = lab_enc.fit_transform(old_artists)

old_dfs['artists'] = encoded

new_artists = new_dfs["artists"]
encoded = lab_enc.fit_transform(new_artists)

new_dfs['artists'] = encoded

#The same for decades
for dec in ALL_dfs.keys():
    artists = ALL_dfs[dec]["artists"]
    encoded = lab_enc.fit_transform(artists)
    ALL_dfs[dec]['artists'] = encoded
```

```python
old_dfs.index.name = "ID"
old_dfs.to_csv("old_VS_new/old_era_data.csv")

new_dfs.index.name = "ID"
new_dfs.to_csv("old_VS_new/new_era_data.csv")
```

```python
for dec in ALL_dfs.keys():
    ALL_dfs[dec].index.name = "ID"
    ALL_dfs[dec].to_csv(f"decades_new/data_from_{dec}s.csv")
```
